## Generacion de variables

```python
import  string
punctuation = string.punctuation
punctuation
```

```python
data['count_punctuation'] = data['verified_reviews'].apply(lambda x: len("". join(_ for _ in x if _ in punctuation)))
```

```python
data['count_char'] = data['verified_reviews'].apply(len)
```

```python
palabra = 'Hola 801, como estan?'
```

```python
palabra.split()
```

```python
data['count_word'] = data['verified_reviews'].apply(lambda x: len(x.split()))
```

```python
data['density'] = data['count_char'] / (data['count_word'] + 1)
```

## Polarizacion 

```python

from textblob import TextBlob
```

```python
def c(x):
    return x*x
```

```python
data['rr'] = data['rating'].apply(c)
```

```python
data.head()
```

```python
testimonial =TextBlob("Textblob is amazingly simple to use. What great fun!")
```

```python
testimonial.sentiment.polarity
```

```python
def pol (text):
    text_blob =TextBlob(str(text.encode('utf-8')))
    pol = text_blob.sentiment.polarity
    return pol
```

```python
texto = ("Textblob is amazingly simple to use. What great fun!")
```

```python
pol(texto)
```

```python
data['polarity'] = data['verified_reviews'].apply(pol)
```

```python
data.head(20)
```

```python
data.sort_values(by=['polarity'])
```

```python
data[data['polarity']<0.3]
```
