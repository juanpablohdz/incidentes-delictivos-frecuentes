 # (a) Â¿Cual es el numero de clientes por region?

```python
df2 = dfclientes[['Region','Customer','City']]
```

```python
dfclientes.count()
```

```python
clientes_region=df2.groupby(['Region'],as_index=False).size().sort_values('size',ascending=False)
clientes_region
```

# (b) Â¿Cual es la region que registra mayores ventas?

```python
ventas=dftransaccional['Sales_Volume'].unique()
ventas
```

```python

dftransaccional['Sales_Volume']=dftransaccional['Sales_Volume'].replace (".", "")
dftransaccional
```

```python
concatenado=pd.concat([dftransaccional, dfclientes], axis=1)
concatenado
```

```python
df7 = concatenado[['Region','Sales_Volume']]
```

```python
clintes_ventas = df7.groupby(['Region','Sales_Volume'],as_index=False).size().sort_values('size',ascending=False)
clintes_ventas
```

# observamos que en la region del sur hay mas ventas.
